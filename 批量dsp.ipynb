```python
import cv2
```

```python

from PIL import Image
import os.path
import glob
def convertpng(pngfile,outdir,width=128,height=128):
  
  image = cv2.imread(pngfile)    
  h_flip = cv2.flip(image, 1)
  cv2.imwrite(os.path.join(outdir,os.path.basename(pngfile)),h_flip)

for i in range (1,33):
  outdir = "/content/dsp_flipping2/batch%03d"%(i)  
  if not os.path.exists(outdir):
        os.makedirs(outdir)
        for pngfile in glob.glob("/content/batchwrong/content/batchwrong/batch%03d"%(i)+"/*.png"):
          convertpng(pngfile,outdir)
```
